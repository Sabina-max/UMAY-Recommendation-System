```python
# =====================================================================================
# # Recommendation System Prototype for UMAY Shop
# 
# ## Goal
# Develop a rule-based recommendation system using publicly available product data.
# 
# ## Data source
# Public API: https://endpoint.umayshop.kz/product/public/web/recommend/product/list
# 
# ## Limitation
# The API provides a recommendation feed rather than the full product catalog.
# Therefore, recommendations are generated within the available public dataset.
# =====================================================================================

import requests
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns

# Pandas output configuration
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# =====================================================================================
# SYSTEM CONFIGURATION & WEIGHTS
# Coefficients are determined expertly based on the implied business value of features.
# In a production environment, weights should be optimized via A/B testing or Optuna.
# =====================================================================================
REQUEST_DELAY = 0.5  # API request delay to prevent rate limiting

MAIN_PAGE_WEIGHTS = {
    "sales": 0.35,       # Strongest indicator of product demand and user trust
    "rating": 0.30,      # High rating reflects product quality and minimizes bad UX
    "discount": 0.20,    # Visual trigger to increase Click-Through Rate (CTR)
    "promotion": 0.15,   # Platform business goals / marketing campaigns priority
    "similarity": 0.00   
}

CART_PAGE_WEIGHTS = {
    "sales": 0.25,       
    "rating": 0.20,      
    "discount": 0.35,    # Maximized to serve as a financial trigger for impulse buying
    "promotion": 0.20,   
    "similarity": 0.00   
}

PRODUCT_PAGE_WEIGHTS = {
    "sales": 0.05,       # Minimized to prioritize contextual relevance over popularity
    "rating": 0.15,      
    "discount": 0.10,    
    "promotion": 0.10,   
    "similarity": 0.60   # Maximum priority for category, price, and keyword matching
}

CHILD_KEYWORDS = ["детск", "ребен", "для детей", "малыш"]

print("Step 1: Data ingestion via UMAY Shop API...")

# ==========================================
# 1. DATA INGESTION
# ==========================================
url = "https://endpoint.umayshop.kz/product/public/web/recommend/product/list"
all_products = []

for page in range(1, 11):
    params = {"pageNo": page, "pageSize": 50}
    try:
        response = requests.get(url, params=params)
        items = response.json()["data"]["list"]
        all_products.extend(items)
        print(f"  Page {page}: fetched {len(items)} items")
    except Exception as e:
        print(f"  Error on page {page}: {e}")
        
    time.sleep(REQUEST_DELAY)

df_raw = pd.DataFrame(all_products)
print(f"Total raw records loaded: {len(df_raw)}")


# ==========================================
# 2. DATA CLEANING & FEATURE ENGINEERING
# ==========================================
print("\nStep 2: Data cleaning and processing...")

# Deduplication by ID and product names to eliminate visual spam
clean_df = df_raw.drop_duplicates(subset=['productId']).copy()
clean_df = clean_df.drop_duplicates(subset=['subject']).copy()

# Robust extraction of the root category from text tags
def extract_root_keyword(text):
    if pd.isna(text) or not isinstance(text, str):
        return "Другое"
    return text.split(':')[0].strip() if ":" in text else text.strip()

clean_df['root_keyword'] = clean_df['recommendKeywords'].apply(extract_root_keyword)

# Mathematical discount percentage calculation
clean_df['discount_percent'] = np.where(
    clean_df['underlinedPrice'] > clean_df['price'],
    ((clean_df['underlinedPrice'] - clean_df['price']) / clean_df['underlinedPrice']) * 100,
    0
)

# Convert promotion flags to binary format
clean_df['promo_score'] = clean_df['isPromotion'].astype(int)

# Filter technical anomalies
clean_df = clean_df[clean_df['price'] > 0]

print(f"Cleaned dataset shape: {clean_df.shape[0]} unique products.")


# ==========================================
# 3. EXPLORATORY DATA ANALYSIS (EDA)
# ==========================================
print("\nStep 3: Visualizing catalog feature distributions...")
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# 1. Macro-categories distribution
top_categories = clean_df['root_keyword'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette="viridis", ax=axes[0, 0])
axes[0, 0].set_title("Top 10 Macro-Categories (via tags)", fontsize=11, fontweight='bold')
axes[0, 0].set_xlabel("Product Count")

# 2. Price distribution
sns.histplot(clean_df['price'], bins=20, kde=True, color="purple", ax=axes[0, 1])
axes[0, 1].set_title("Price Distribution (KZT)", fontsize=11, fontweight='bold')
axes[0, 1].set_xlabel("Price")
axes[0, 1].set_ylabel("Count")

# 3. Sales volume distribution
sns.histplot(clean_df['sales'], bins=15, kde=True, color="teal", ax=axes[1, 0])
axes[1, 0].set_title("Sales Volume Distribution (sales)", fontsize=11, fontweight='bold')
axes[1, 0].set_xlabel("Sales Units")
axes[1, 0].set_ylabel("Card Count")

# 4. User rating distribution
sns.histplot(clean_df['showScore'], bins=10, kde=False, color="orange", ax=axes[1, 1])
axes[1, 1].set_title("Customer Rating Distribution (showScore)", fontsize=11, fontweight='bold')
axes[1, 1].set_xlabel("Rating Value")
axes[1, 1].set_ylabel("Card Count")

plt.tight_layout()
plt.show()


# ==========================================
# 4. MULTI-CRITERIA SCORING LOGIC
# ==========================================
def calculate_weighted_score(dataframe, weights, target_category_id=None, target_price=None, target_keyword=None):
    """
    Computes a linear weighted ranking score using isolated MinMax normalization.
    """
    df_scored = dataframe.copy()
    
    # Bugfix: Normalizing strictly within the local passed dataframe boundaries
    df_scored['sales_norm'] = (df_scored['sales'] - df_scored['sales'].min()) / (df_scored['sales'].max() - df_scored['sales'].min() + 1e-5)
    df_scored['score_norm'] = (df_scored['showScore'] - df_scored['showScore'].min()) / (df_scored['showScore'].max() - df_scored['showScore'].min() + 1e-5)
    df_scored['discount_norm'] = (df_scored['discount_percent'] - df_scored['discount_percent'].min()) / (df_scored['discount_percent'].max() - df_scored['discount_percent'].min() + 1e-5)
    
    # Calculate contextual similarity for Product Page scenario
    if target_category_id is not None and target_price is not None and target_keyword is not None:
        df_scored['sim_category'] = (df_scored['secondCategoryId'] == target_category_id).astype(int)
        df_scored['sim_keyword'] = (df_scored['root_keyword'] == target_keyword).astype(int)
        
        # Gaussian price proximity metric
        df_scored['sim_price'] = np.exp(-((df_scored['price'] - target_price) / (target_price * 0.4))**2)
        
        # Combined similarity index
        df_scored['similarity'] = 0.5 * df_scored['sim_category'] + 0.3 * df_scored['sim_price'] + 0.2 * df_scored['sim_keyword']
    else:
        df_scored['similarity'] = 0.0
        df_scored['sim_category'] = 0
        df_scored['sim_price'] = 0.0
        
    # Final weighted ranking score computation
    df_scored['final_score'] = (
        weights["sales"] * df_scored['sales_norm'] +
        weights["rating"] * df_scored['score_norm'] +
        weights["discount"] * df_scored['discount_norm'] +
        weights["promotion"] * df_scored['promo_score'] +
        weights["similarity"] * df_scored['similarity']
    )
    
    return df_scored.sort_values(by='final_score', ascending=False)


# ==========================================
# 5. RECOMMENDATION EXPLANATION
# ==========================================
def generate_objective_explanation(row, is_card_scenario=False):
    """
    Generates transparent technical reasons for product recommendation.
    """
    reasons = []
    if row['sales'] >= 10000:
        reasons.append(f"Высокие продажи ({int(row['sales'])}+ шт.)")
    elif row['sales_norm'] > 0.7:
        reasons.append(f"Стабильный спрос ({int(row['sales'])} шт.)")
        
    if row['showScore'] >= 4.5:
        reasons.append(f"Высокий рейтинг ({row['showScore']:.1f}/5.0)")
        
    if row['discount_percent'] > 0:
        reasons.append(f"Действующая скидка ({row['discount_percent']:.1f}%)")
        
    if row['promo_score'] == 1:
        reasons.append("Участник промо-акции")
    
    if is_card_scenario:
        if row.get('sim_category', 0) == 1:
            reasons.append("Совпадение подкатегории")
        if row.get('sim_price', 0) > 0.8:
            reasons.append("Ценовое сходство")
            
    return " | ".join(reasons) if reasons else "Базовое ранжирование по каталогу"


# ==========================================
# 6. DIVERSIFICATION (INTERLEAVING)
# ==========================================
def apply_diversification(scored_dataframe, top_n=10, category_col='root_keyword'):
    """
    Interleaves categories to ensure continuous type alternation on the storefront.
    """
    # Bugfix: Ensuring dataframe is sorted by final_score before grouping
    sorted_df = scored_dataframe.sort_values(by='final_score', ascending=False)
    chunks = [frame for _, frame in sorted_df.groupby(category_col, sort=False)]
    
    result = []
    for i in range(len(sorted_df)):
        added_in_this_round = 0
        for chunk in chunks:
            if i < len(chunk):
                result.append(chunk.iloc[i])
                added_in_this_round += 1
        if added_in_this_round == 0:
            break
    return pd.DataFrame(result).head(top_n)


# ==========================================
# 7. BUSINESS RECOMMENDATION STRATEGIES
# ==========================================

# Scenario 1: Main Page Strategy (Diversified Ranking)
def get_main_page_recommendations(dataframe, top_n=10):
    scored = calculate_weighted_score(dataframe, weights=MAIN_PAGE_WEIGHTS)
    diversified = apply_diversification(scored, top_n=top_n, category_col='root_keyword')
    diversified['Обоснование рекомендации'] = diversified.apply(lambda r: generate_objective_explanation(r), axis=1)
    
    cols = ['productId', 'subject', 'root_keyword', 'price', 'discount_percent', 'showScore', 'final_score', 'Обоснование рекомендации']
    return diversified[cols]

# Scenario 2: Product Page Strategy (Context-Aware Mapping with Content Isolation)
def get_product_page_recommendations(dataframe, current_product_id, top_n=10):
    target = dataframe[dataframe['productId'] == current_product_id]
    if target.empty: 
        return "Product ID not found in current dataset boundary."
    
    t_row = target.iloc[0]
    t_cat = t_row['secondCategoryId']
    t_price = t_row['price']
    t_key = t_row['root_keyword']
    t_sub = t_row['subject'].lower()
    
    filtered = dataframe[dataframe['productId'] != current_product_id].copy()
    
    pattern = "|".join(CHILD_KEYWORDS)
    is_target_child = any(word in t_sub for word in CHILD_KEYWORDS)
    
    if is_target_child:
        filtered = filtered[filtered['subject'].str.lower().str.contains(pattern, regex=True)]
    else:
        filtered = filtered[~filtered['subject'].str.lower().str.contains(pattern, regex=True)]
        
    scored = calculate_weighted_score(filtered, weights=PRODUCT_PAGE_WEIGHTS, 
                                      target_category_id=t_cat, target_price=t_price, target_keyword=t_key)
    
    scored['Обоснование рекомендации'] = scored.apply(lambda r: generate_objective_explanation(r, is_card_scenario=True), axis=1)
    
    cols = ['productId', 'subject', 'root_keyword', 'price', 'discount_percent', 'final_score', 'Обоснование рекомендации']
    return scored.head(top_n)[cols]

# Scenario 3: Empty Cart Strategy (Financial Trigger Drive)
def get_empty_cart_recommendations(dataframe, top_n=10):
    scored = calculate_weighted_score(dataframe, weights=CART_PAGE_WEIGHTS)
    diversified = apply_diversification(scored, top_n=top_n, category_col='root_keyword')
    diversified['Обоснование рекомендации'] = diversified.apply(lambda r: generate_objective_explanation(r), axis=1)
    
    cols = ['productId', 'subject', 'root_keyword', 'price', 'discount_percent', 'final_score', 'Обоснование рекомендации']
    return diversified[cols]


# ==========================================
# 8. PROCESS INFRASTRUCTURE & REVIEWS
# ==========================================
print("\nStep 4: Executing recommendation pipeline pipelines...")

print("\n" + "="*110)
print(" 📋 SCENARIO 1: MAIN PAGE RECOMMENDATIONS (Diversified Category Interleaving)")
print("="*110)
display(get_main_page_recommendations(clean_df))

sample_id = '1905777800969281538'
print("\n" + "="*110)
print(f" 👕 SCENARIO 2: PRODUCT PAGE CONTEXT ANALOGS (Target ID: {sample_id} | Content Safety Filter)")
print("="*110)
display(get_product_page_recommendations(clean_df, current_product_id=sample_id))

print("\n" + "="*110)
print(" 🛍️ SCENARIO 3: EMPTY CART RECOMMENDATIONS (Financial Triggers Focus)")
print("="*110)
display(get_empty_cart_recommendations(clean_df))
```
